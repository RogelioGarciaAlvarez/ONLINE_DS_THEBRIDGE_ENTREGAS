```python
import matplotlib.pyplot as plt
import numpy as np
```

### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
#Creamos los valores de x y las funciones seno y coseno de x
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

#Creamos la figura con los dos subplots
plt.subplot(1, 2, 1)  #1 fila, 2 columnas, primer subplot
plt.plot(x, y1)
plt.title('Seno')

plt.subplot(1, 2, 2)  #1 fila, 2 columnas, segundo subplot
plt.plot(x, y2)
plt.title('Coseno')

plt.tight_layout() #Para que todo se presente limpio
plt.show()
```

### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

```python
#Creamos la funcion tangente de x
x = np.linspace(0, 10, 100)
y3 = np.tan(x)

#Creamos la figura con los tres subplots
fig, ax = plt.subplots(3,1)  #1 fila, 3 columnas

ax[0].plot(x, y1, color = "red") #Primer subplot
ax[1].plot(x, y2, color = "green") #Segundo subplot
ax[2].plot(x, y3, color = "blue") #Tercer subplot

plt.tight_layout() #Para que todo se presente limpio
plt.show()
```


### Ejercicio 3: Ajuste de Colores y Estilos de Lnea  

Modifica la primera figura para que se muestre de la siguiente manera

```python
#Creamos las funciones seno y coseno de 2x
y1_2 = np.sin(2*x)
y2_2 = np.cos(2*x)

#Creamos la figura con los dos subplots
fig, ax = plt.subplots(1,2)  #1 fila, 2 columnas

ax[0].plot(x, y1_2, linestyle = "--", color = "magenta") #Primer subplot
ax[1].plot(x, y2_2, linestyle = "-.", color = "cyan") #Segundo subplot

plt.tight_layout() #Para que todo se presente limpio
plt.show()
```

### Ejercicio 4

Reproduce la figura siguiente. Recuerda que las funciones seno hiperblico y coseno hiperblico se obtienen as:
* np.sinh, seno hiperblico;
* np.cosh, coseno hiperblico;


```python
#Creamos las funciones seno hiperbolico y coseno hiperbolico de x
y4 = np.sinh(x)
y5 = np.cosh(x)


#Creamos la figura con los dos subplots
fig, ax = plt.subplots(2,1) #2 filas, 1 columna

ax[0].plot(x, y4, color = "red") #Primer subplot
ax[0].set_title("Seno Hiperblico")
ax[0].set_xlabel("x")
ax[0].set_ylabel("sinh(x)")

ax[1].plot(x, y5, color = "blue") #Segundo subplot
ax[1].set_title("Coseno Hiperblico")
ax[1].set_xlabel("x")
ax[1].set_ylabel("cosh(x)")

plt.tight_layout() #Para que todo se presente limpio
plt.show()
```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosicin (tal y como ves en la figura), emplea `subplots_adjust` y los parmetros hspace y wspace.

```python
#Creamos la funcion arcotangente de x
y6 = np.arctan(x)

#Creamos la figura con los dos subplots
fig, ax = plt.subplots(2,2) #2 filas, 2 columnas
fig.subplots_adjust(hspace = 0.5, wspace = 0.5) #Ajusta el espaciado

ax[0,0].plot(x, y1) #Primer subplot
ax[0,1].plot(x, y2) #Segundo subplot
ax[1,0].plot(x, y3) #Tercer subplot
ax[1,1].plot(x, y6) #Cuarto subplot

plt.show()
```

Reproduce la figura siguiente empleando slo el estilo MATLAB (plt.\<funcion>\) y creando solo el contenido de los dos primeros grficos (fila 1). Luego usa el estilo con mtodos (objetos `axes`) para aadir el tercer grfico.
