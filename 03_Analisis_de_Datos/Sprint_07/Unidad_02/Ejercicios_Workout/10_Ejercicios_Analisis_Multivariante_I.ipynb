* Existe alguna diferencia visualmente significativa entre las propinas de los grupos de la variable "sex"?
* Son diferentes respecto al  conjunto total?

```python
#obtengo las medidas numricas
df_4.groupby("sex")["tip"].describe()
```

```python
#Calculo el Coeficiente de Variacion
for valor in df_4["sex"].unique():
    cv = (df_4.loc[df_4["sex"] == valor,"tip"].describe().T["std"]/df_4.loc[df_4["sex"] == valor, "tip"].describe().T["mean"])*100
    cv = round(cv,2)
    print(f"El Coeficiente de Variacin de <tip> para {valor} es {cv}%")
```

Es una variable con cierta dispersin, por lo que me quedo con la mediana para caracterizar al conjunto de datos

### Ejercicio 9

Usa scipy tal como lo hicimos en el workout para hacer el test de correlacin. Cul es el p-value?Qu podemos decir de la correlacin lineal entre "total" y "speeding"?

```python
#Realizo el test de correlacin lineal
pearsonr(df_7["speeding"], df_7["total"])
```

Un p-value por debajo de 0.05 rechaza la hiptesis nula (no hay relacin lineal entre ambas variables)

El p-value es muy pequeo, se puede rechazar la hipotesis de independencia de las variables y decir que existe una relacin lineal entre "total" y "speeding" estadsticamente significativa.
