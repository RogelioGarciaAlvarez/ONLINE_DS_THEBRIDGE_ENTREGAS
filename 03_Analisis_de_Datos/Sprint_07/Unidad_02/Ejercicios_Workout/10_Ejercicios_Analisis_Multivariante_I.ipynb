Haz el análisis bivariante de "tip" y "sex" (emplea las funciones del workout importándolas del módulo `bootcampviztools`, o utiliza las visualizaciones del notebook en el repositorio), añade el cálculo de la media y mediana de la variable continua por cada valor de la variable categórica, y contesta a las siguientes preguntas:
* ¿Existe alguna diferencia visualmente significativa entre las propinas de los grupos de la variable "sex"?
* ¿Son diferentes respecto al  conjunto total?

```python
#obtengo las medidas numéricas
df_4.groupby("sex")["tip"].describe()
```

```python
#Calculo el Coeficiente de Variacion
for valor in df_4["sex"].unique():
    cv = (df_4.loc[df_4["sex"] == valor,"tip"].describe().T["std"]/df_4.loc[df_4["sex"] == valor, "tip"].describe().T["mean"])*100
    cv = round(cv,2)
    print(f"El Coeficiente de Variación de <tip> para {valor} es {cv}%")
```

Es una variable con cierta dispersión, por lo que me quedo con la mediana para caracterizar al conjunto de datos

```python
#Obtengo la distribucion global de "sex" para comparar
viz.pinta_distribucion_categoricas(df_4, ["sex"], relativa=True, mostrar_valores=True)
```

```python
#Comparo la mediana de "tip" para cada valor de "sex"
viz.plot_categorical_numerical_relationship(df_4,categorical_col="sex",numerical_col="tip", show_values=True, measure="median")
```

```python
#Obtengo la frecuencia, función de distibución y digrama de caja de "tip" para comparar
viz.plot_combined_graphs(df_4, ["tip"], whisker_width=1.5, bins=20)
```

```python
#Comparo los diagramas de caja para valor de "sex"
viz.plot_grouped_boxplots(df_4,cat_col="sex", num_col="tip", group_size=2)
```

```python
#Comparo los histogramas para cada valor de "sex"
viz.plot_grouped_histograms(df_4, cat_col="sex", num_col="tip",group_size=2)
```

* Sí hay diferencia, para el valor "Male" (hombres) la mediana de las propinas es superior a la mediana para valor "F" (mujeres) y comparando los histogramas también se ven valores mas dispersos y mayor rango,
por lo que parece que los hombres tienden a dejar propina más altas

* La distribucion de "Male" es muy parecida a la distribución total debido a la mayor presencia de hombres (esto influye en la distribución total). Esto sugiere un analisis por separado de hombres y mujeres

### Ejercicio 9

Usa scipy tal como lo hicimos en el workout para hacer el test de correlación. ¿Cuál es el p-value?¿Qué podemos decir de la correlación lineal entre "total" y "speeding"?

```python
#Realizo el test de correlación lineal
pearsonr(df_7["speeding"], df_7["total"])
```

Un p-value por debajo de 0.05 rechaza la hipótesis nula (no hay relación lineal entre ambas variables)

El p-value es muy pequeño, se puede rechazar la hipotesis de independencia de las variables y decir que existe una relación lineal entre "total" y "speeding" estadísticamente significativa.
